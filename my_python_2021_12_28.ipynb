## conda基本操作

conda list ```查看已安装的软件包```

conda install package_name ```下载安装包```

conda unistall package_name

conda create -n tensorflow35（环境名） python=3.5(版本名）      ```用于conda创建环境```

activate tensorflow       ```打开已创建的环境```  

deactivate ```退出环境```

conda env list        ```查看安装的环境```

conda remove -n rcnn --all ```删除环境```

conda create -n tf --clone rcnn ```复制环境```

````python
import sklearn.datasets as datasets
import pandas as pd
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
````

````python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
````

````python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
````

## 矩阵

````python
import numpy as np      #引入numpy
A = np.array([1,2,3])   #转为矩阵
B = np.array([4,3,0])
print('A+B = ',np.add(A,B))
print('A-B = ',np.subtract(A,B))
````

````python
C = np.array([[1,2,3],
            [1,1,1]])
D = np.array([[4,3,9,9],
            [1,1,2,2,],
            [4,2,1,1]])
#矩阵相乘
print('np.dot(C,D) = ',np.dot(C,D))
print('np.matmul(C,D) = ',np.matmul(C,D))
print('        C@D    = ', C@D)   

#元素与元素相乘
print('np.multiply(D,D)=',np.multiply(D,D))
print('          D*D   = ', D*D)
````

````python
X = np.array([[1],
            [2],
            [3]])
Y = np.array([[1,2,3]])
H = np.dot(X,Y)
I = X@Y
````
